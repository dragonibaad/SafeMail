```python

import pandas as pd
import re
df = pd.read_csv('./Email_Dataset.csv')


def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    text = re.sub(r"[^a-zA-Z\s]", " ", text)
    text = re.sub(r"\s+", " ", text)
    return text.strip()

df['clean'] = df['Email Text'].apply(clean_text)
df.head()
df.to_csv('Clean_Proj_Dataset.csv')
```
